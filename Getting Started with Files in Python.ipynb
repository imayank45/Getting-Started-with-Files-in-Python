# Text Preprocessing

```python
paragraph=txt.split('\n\n')
len(paragraph)
```

```python
for i in paragraph:
    print(i)
```

```python
lines=txt.split('.')
len(lines)
```

```python
words=txt.split(' ')
len(words)
```

```python
words
```

# Advance Text Processing

```python
print(txt)
```

```python
for i in range(1,27):
    pattern='['+str(i)+']'
    txt=txt.replace(pattern,'')
```

```python
print(txt)
```

```python
for i in '!@#$%^&*()_+':
    txt=txt.replace(i,'')
print(txt)
```
